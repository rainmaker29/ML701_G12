# Testing to see if the features have normal distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
%cd "Desktop"
data = pd.read_csv('dataset.csv')

#features of interest
df=data[["u", "g",'r','i','z']]
```

```python
from sklearn.preprocessing import StandardScaler 
scale = StandardScaler()

data_scaled=scale.fit_transform(df)
df_scaled = pd.DataFrame(data_scaled, columns = ['u','g','r','i','z'])
```

```python
r =df_scaled.loc[:,"r"]
i =df_scaled.loc[:,"i"]
z =df_scaled.loc[:,"z"]
u =df_scaled.loc[:,"u"]
g =df_scaled.loc[:,"g"] 
```

#           Histogram test for scaled features

```python
fig = plt.figure(figsize=(20, 18))
ax = fig.add_subplot()
num=50
_ = ax.hist(u.values, color='green', alpha=.3, bins=num)
_ = ax.hist(g.values, color='black', alpha=.3, bins=num)
_ = ax.hist(r.values, color='blue',alpha=.3, bins=num)
_ = ax.hist(i.values, color='red', alpha=.3, bins=num)
_ = ax.hist(z.values, color='yellow', alpha=.3, bins=num)
plt.legend(['u','g','r','i','z'], loc ="lower right", fontsize=20)
plt.xlim([-5,5])
```

   #        QQ Plot test
   QQ Plot stands for Quantile vs Quantile Plot: it plots theoretical quantiles against the actual     quantiles of our variable.
   
   Quantiles are cut points dividing the range of a probability distribution into continuous intervals with equal probabilities, or dividing the observations in a sample in the same way:
   


   #####                                                  Normal (left), uniform (middle) and exponential (right) QQ Plots

```python
from statsmodels.graphics.gofplots import qqplot 
from matplotlib import pyplot as plt 

qqplot(r,fit=True,line="45") 


plt.show()
```

```python
qqplot(i,fit=True,line="45") 
plt.show()
```

```python
qqplot(z,fit=True,line="45") 
plt.show()
```
